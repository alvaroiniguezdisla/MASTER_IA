# Reto Machine Learning III - Análisis de Componentes

PCA sobre el dataset de precios de casas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("/workspace/dataset.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum().sort_values(ascending=False).head(20)
```

```python
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns
categorical_cols = df.select_dtypes(include=["object"]).columns

print("Columnas numéricas:", len(numeric_cols))
print("Columnas categóricas:", len(categorical_cols))
```

```python
df_clean = df.copy()

for col in numeric_cols:
    df_clean[col] = df_clean[col].fillna(df_clean[col].median())

for col in categorical_cols:
    df_clean[col] = df_clean[col].fillna(df_clean[col].mode()[0])

df_clean.isnull().sum().sum()
```

```python
df_encoded = pd.get_dummies(df_clean, columns=categorical_cols, drop_first=False)
df_encoded.head()
```

```python
X_scaled = StandardScaler().fit_transform(df_encoded)
```

```python
pca = PCA(n_components=30, random_state=42)
X_pca = pca.fit_transform(X_scaled)

print("Forma original:", df_encoded.shape)
print("Forma transformada:", X_pca.shape)
```

```python
explained_variance = pca.explained_variance_ratio_
cumulative_variance = explained_variance.cumsum()

variance_df = pd.DataFrame({
    "Componente": range(1, 31),
    "Varianza explicada": explained_variance,
    "Varianza acumulada": cumulative_variance
})

variance_df.head()
```

```python
plt.figure(figsize=(12, 6))
components = range(1, 31)

plt.bar(components, explained_variance, alpha=0.7, label="Varianza explicada")
plt.plot(components, cumulative_variance, marker="o", label="Varianza acumulada")

plt.xlabel("Componente principal")
plt.ylabel("Proporción de varianza")
plt.title("PCA: varianza explicada y varianza acumulada")
plt.xticks(components)
plt.legend()
plt.grid(axis="y", alpha=0.3)
plt.show()
```

```python
print("Varianza explicada acumulada por los 30 componentes:")
print(cumulative_variance[-1])
```

```python
if cumulative_variance[-1] >= 0.80:
    print("La reducción a 30 componentes retiene una gran parte de la información del dataset.")
else:
    print("La varianza acumulada es inferior al 80%, aunque para este proyecto se mantiene PCA con 30 componentes.")
```
